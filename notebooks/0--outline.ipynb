````python
from IPython.display import Markdown, display
import textwrap

def show_doc(obj):
    """Nicely display a docstring using Markdown in Jupyter."""
    doc = obj.__doc__
    if doc:
        display(Markdown(f"```python\n{textwrap.dedent(doc).strip()}\n```"))
    else:
        display(Markdown("_No docstring found._"))

from local_nbutils.utils import load_data
type(load_data)
show_doc(load_data)
````
